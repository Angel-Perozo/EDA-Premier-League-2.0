```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 1.- Carga de la data

```python
data = pd.read_csv('./eplmatches.csv')

data.head()
```

```python
# Conversión de los tipos de datos a formatos más eficientes

data['Date'] = pd.to_datetime(data['Date'])

data['FTR'] = data['FTR'].astype('category')
```

```python
data.info()
```

## 2.- Creación de nueva data para facilitar el entendimiento de nuestros datos

```python
# Para conocer el total de goles anotados por partido

data['TotalGoals'] = data['HomeGoals'] + data['AwayGoals']
```

```python
data['WinningTeam'] =  [data['Home'].values[i] if data['FTR'].values[i] == 'H' else  
                        data['Away'].values[i] if data['FTR'].values[i] == 'A' else 
                        np.nan for i in range(len(data))]

data['LosingTeam'] =   [data['Home'].values[i] if data['FTR'].values[i] == 'A' else 
                        data['Away'].values[i] if data['FTR'].values[i] == 'H' else 
                        np.nan for i in range(len(data))] 

# En busca de entender mejor nuestra data creamos las columnas ['WinningTeam'] & ['LosingTeam'] 
```

```python
data['Result'] = data['HomeGoals'].astype(str) + ' - ' + data['AwayGoals'].astype(str)

# La columna ['Result'] viene de la concatenación de 2 columnas de tipo -int- que fueron previamente llevadas a tipo -str-
# por lo que su uso es primeramente de tipo gráfico por qué no será posible iterar sobre ella
```

```python
data.sort_values(by = ['Date'], ascending = True, ignore_index = True, inplace = True)

# Nos aseguramos que los partidos estén organizados en base a la fecha
```

## 5 .- Goles Anotados vs Recibidos

```python
# Goles Anotados

HomeGoalCount = data.groupby(['Home'])['HomeGoals'].sum().reset_index()
HomeGoalCount.columns = ['index','Home Goals']

AwayGoalCount = data.groupby(['Away'])['AwayGoals'].sum().reset_index()
AwayGoalCount.columns = ['index','Away Goals']

TotalGoalScored = pd.merge(HomeGoalCount, AwayGoalCount, how ='right', on ='index')
TotalGoalScored.sort_values(by=['Home Goals','Away Goals'], ascending = False).head(10)
```

```python
TotalGoalScored['Total'] = TotalGoalScored['Home Goals'] + TotalGoalScored['Away Goals']
TotalGoalScored['Diferencial'] = TotalGoalScored['Home Goals'] - TotalGoalScored['Away Goals']
TotalGoalScored.sort_values(by = ['Total'], ascending = False).head(10)
```

```python
Top10Goals = TotalGoalScored.sort_values(by=['Total'], ascending = False).head(10)
Top10Goals = Top10Goals.drop('Total', axis = 1)
```

```python
Top10Goals.plot(x='index', kind='bar', width = 0.8)
plt.tick_params(axis='both', labelsize=20)

plt.title('Home vs Away Goals Scored', fontsize = 30, fontweight = 'bold')
plt.xlabel('Teams', fontsize = 30, fontweight = 'bold')
plt.ylabel('Goals', fontsize = 30, fontweight = 'bold')
plt.show()
```

```python
# Goles Concedidos 

HomeGoalConceded = data.groupby(['Home'])['AwayGoals'].sum().to_frame().reset_index().rename({'Home' : 'index'}, axis = 1)

AwayGoalConceded = data.groupby(['Away'])['HomeGoals'].sum().to_frame().reset_index().rename({'Away' : 'index'}, axis = 1)

TotalGoalConceded = pd.merge(AwayGoalConceded, HomeGoalConceded, how = 'right', on = 'index' )
TotalGoalConceded.sort_values(by = ['HomeGoals', 'AwayGoals'], ascending = False).head(10)
```

```python
TotalGoalConceded['Total'] = TotalGoalConceded['HomeGoals'] + TotalGoalConceded['AwayGoals']
TotalGoalConceded['Diferencial'] = TotalGoalConceded['HomeGoals'] - TotalGoalConceded['AwayGoals']
TotalGoalConceded.sort_values(by = ['Total'], ascending = False).head(10)
```

```python
Top10GoalsConced = TotalGoalConceded.sort_values(by=['Total'], ascending = False).head(10)
Top10GoalsConced = Top10GoalsConced.drop('Total', axis = 1)
```

```python
Top10GoalsConced.plot(x='index', kind='bar',width = 0.8)
plt.tick_params(axis='both', labelsize=20)

plt.title('Home vs Away Goals Conceded', fontsize = 30, fontweight = 'bold')
plt.xlabel('Teams', fontsize = 30, fontweight = 'bold')
plt.ylabel('Goals', fontsize = 30, fontweight = 'bold')
plt.show()
```

```python
# Goles anotados por temporada y por equipo

HomeGoalScoredPerSeason = data.groupby(['Home', 'Season_End_Year'])['HomeGoals'].sum().reset_index()
HomeGoalScoredPerSeason.columns = ['Team', 'Season', 'HGSPS']

AwayGoalScoredPerSeason = data.groupby(['Away', 'Season_End_Year'])['AwayGoals'].sum().reset_index()
AwayGoalScoredPerSeason.columns = ['Team', 'Season', 'AGSPS']

TotalGoalScoredPerSeason = pd.merge(HomeGoalScoredPerSeason, AwayGoalScoredPerSeason, how='right', on=['Team', 'Season'])

TotalGoalScoredPerSeason
```

```python
# Goles concedidos por temporada y por equipo

HomeGoalConcededPerSeason = data.groupby(['Home', 'Season_End_Year'])['AwayGoals'].sum().reset_index()
HomeGoalConcededPerSeason.columns = ['Team', 'Season', 'HGCPS']

AwayGoalConcededPerSeason = data.groupby(['Away', 'Season_End_Year'])['HomeGoals'].sum().reset_index()
AwayGoalConcededPerSeason.columns = ['Team', 'Season', 'AGCPS']

TotalGoalConcededPerSeason = pd.merge(HomeGoalConcededPerSeason, AwayGoalConcededPerSeason, how='right', on=['Team', 'Season'])

TotalGoalConcededPerSeason
```

```python
TotalGoalsPerSeason = pd.merge(TotalGoalScoredPerSeason, TotalGoalConcededPerSeason, how='right', on=['Team', 'Season'])

TotalGoalsPerSeason
```

```python
TotalGoalsPerSeason['TGSPS'] = TotalGoalsPerSeason['HGSPS'] + TotalGoalsPerSeason['AGSPS']
TotalGoalsPerSeason['TGCPS'] = TotalGoalsPerSeason['HGCPS'] + TotalGoalsPerSeason['AGCPS']
TotalGoalsPerSeason['DGPS'] = TotalGoalsPerSeason['TGSPS'] - TotalGoalsPerSeason['TGCPS']
```

```python
# HGSPS = Home Goals Scored per Season
# AGSPS = Away Goals Scored per Season
# HGCPS = Home Goals Conceded per Season
# AGCPS = Away Goals Conceded per Season
# TGSPS = Total Goals Scored per Season
# TGCPS = Total Goals Conceded per Season
# DGPS = Diferencial Goals per Season

TotalGoalsPerSeason.loc[(TotalGoalsPerSeason['Team'] == 'Sunderland'), :].\
style.highlight_max(['HGSPS', 'AGSPS', 'HGCPS', 'AGCPS', 'TGSPS', 'TGCPS', 'DGPS'])
```

```python
# Agrupamos primero por temporada y victorias 

Win = data.groupby(['Season_End_Year'])['WinningTeam'].value_counts().to_frame().sort_index().rename({'WinningTeam' : 'Victorias'}, axis = 1)
Win.head(10)
```

```python
# Ahora por temporada y derrotas

Lose = data.groupby(['Season_End_Year'])['LosingTeam'].value_counts().to_frame().sort_index().rename({'LosingTeam' : 'Derrotas'}, axis = 1)
Lose.head(10)
```

```python
Historico['Draws'] = 38 - (Historico['Wins'] + Historico['Losses'])
```

```python
# Recordando que las tres primeras temporadas se disputaron con más equipos hacemos la corrección

Historico['Draws'] = np.where(Historico['Season'] <= 1995, (Historico['Draws'] + 4), Historico['Draws'])
```

```python
Historico['Points'] = (Historico['Wins'] * 3) + (Historico['Draws'])
```

```python
# Convertiremos las columnas de la consulta anterior en una Lista que luego uniremos para formar un diccionario que
# facilitará las futuras consultas a realizar, en donde las temporadas representaran las llaves del diccionario y 
# los puntos serán los valores de cada llave

x = list(Champ.Season)
y = list(Champ.Points)
z = dict(zip(x,y))
print(z)
```

```python
def filter_teams(group):
    points = z[group.name]
    teams = group[group['Points'] == points]
    return teams.iloc[0]

result = Historico.groupby('Season').apply(filter_teams)
```

## 11.- Top5 y el Big6

```python
# Añadir la columna 'Position'

Historico = Historico.sort_values(by=['Season', 'Points'], ascending=[True, False])
Historico['Position'] = Historico.groupby('Season').cumcount() + 1

Historico
```

```python
# Podemos ver que se hizo correctamente la asignación de las posiciones por temporada. 
# tenemos 30 registros de numero 1 osea campeón y así sucesivamente, además podemos notar los datos atípicos 
# correspondientes a las posiciones 21 y 22 en 3 oportunidades, con lo que recordamos que las 3 primeras 
# temporadas se disputaron con 22 equipos

Historico['Position'].value_counts()
```

```python
table = pd.pivot_table(Historico, index='Season', columns='Position', values='Team', aggfunc='first').fillna('')

table = table.rename(columns={1: '1ro', 2: '2do', 3: '3ro', 4: '4to', 5: '5to'})

table = table[['1ro', '2do', '3ro', '4to', '5to']]

table = table.reset_index()

table.head(10)
```

```python
v1 = table['1ro'].value_counts()
v2 = table['2do'].value_counts()
v3 = table['3ro'].value_counts()
v4 = table['4to'].value_counts()
v5 = table['5to'].value_counts()

Top5 = pd.concat([v1, v2, v3, v4, v5], axis = 1)

Top5
```

```python
Top5 = Top5.fillna(0)
```

```python
Top5.info()
```

```python
Top5[['1ro', '2do', '3ro', '4to', '5to']] = Top5[['1ro', '2do', '3ro', '4to', '5to']].astype(int)
```

```python
Top5['Top5'] = Top5['1ro'] + Top5['2do'] + Top5['3ro'] + Top5['4to'] + Top5['5to'] 
```

```python
# De nuevo encontramos al Manchester Utd estableciendo marcas dentro de la competición, como la de haber estado en el top5
# de la liga en 26 de las 30 temporadas disputadas, menciones también para Liverpool(20), Chelsea(21) y Arsenal(25)
# Tottenham(12) y Everton(4) completan el grupo de los 6 equipos que han disputado todas las temporadas de Premier League

Top5.style.highlight_max(color = 'green')
```
